# Demonstration of fontdict attribute to set the fonts for xlabel, ylabel, and title

```python
font1 ={'font':'serif','color':'#f11f11','size':10}
font2 ={'font':'serif','color':'#f11f11','size':10}
font2 ={'font':'monospace','color':'#ff1111','size':15}

xlist1 = np.array([12,34,56,78])
ylist1 = np.array([822,844,866,898])
xlist2 = np.array([22,34,46,58])
ylist2 = np.array([982,994,1106,1118])

plt.plot(xlist1,ylist1,'x--r')
plt.xlabel("Age",fontdict=font1)
plt.ylabel("Salary/day (Rs.)",fontdict=font1)
plt.title("Salary of aam aadmi",fontdict=font2,loc='right')
```
